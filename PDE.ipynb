## Persamaan Diferensial Parsial

Untuk meninjau kembali latar belakang dasar yang sangat penting mengenai metode numerik untuk persamaan diferensial parsial (PDE). Sebagian besar yang akan kita gunakan adalah pengetahuan umum tentang klasifikasi PDE, teknik penyelesaian untuk setiap jenis PDE, serta beberapa pemahaman mengenai sifat-sifat solusinya. Hal ini juga mencakup pertanyaan-pertanyaan yang berkaitan dengan well-posedness (keberadaan, keunikan, dan kestabilan solusi), kondisi batas, dan isu-isu lainnya.

Pertama, beberapa notasi untuk memulai, kita akan menggunakan notasi
$$
    \frac{\partial u}{\partial x} = u_x
$$

untuk menyatakan turunan parsial dari fungsi 𝑢 terhadap variabel 𝑥

Turunan tingkat lebih tinggi (orde dua atau lebih) atau turunan campuran bisa ditulis dengan cara yang sama, misalnya:

$$
    \frac{\partial^2 u}{\partial x^2} = u_{xx} \quad \quad \frac{\partial^2 u}{\partial y \partial x} = u_{xy}
$$

Perlu diperhatikan bahwa urutan pengambilan turunan kadang bisa berpengaruh. Artinya, turunan terhadap 
𝑥 lalu 𝑦 belum tentu selalu sama dengan terhadap 𝑦 lalu 𝑥, tergantung kondisi fungsi tersebut.

Kapan urutan turunan boleh ditukar?
Urutan turunan bisa ditukar jika fungsi tersebut cukup halus (kontinu dan memiliki turunan yang baik).

Selain menggunakan variabel dalam koordinat Kartesius (seperti x dan 𝑦), kita juga bisa menggunakan koordinat lain, misalnya koordinat polar. Contohnya:

$$
    \frac{\partial u}{\partial \theta} = u_\theta
$$

artinya turunan fungsi 𝑢 terhadap sudut 𝜃.

Atau terhadap waktu:

$$
    \frac{\partial u}{\partial t} = u_t.
$$  

artinya turunan fungsi 𝑢 terhadap waktu 𝑡.

## 1. Klasifikasi PDE

Pengelompokan (klasifikasi) Persamaan Diferensial Parsial (PDE) biasanya didasarkan pada bentuk umum PDE linear orde dua seperti berikut:
$$
    a u_{xx} + 2b u_{xy} + c u_{yy} + d u_x + e u_y + f u = g.
$$
Perhatikan bahwa dalam kerangka umum untuk klasifikasi, 𝑥 atau 𝑦 juga bisa berupa 𝑡 atau variabel lainnya. Secara formal, dengan mengambil transformasi Fourier dari
$$
    a u_{xx} + 2b u_{xy} + c u_{yy} + d u_x + e u_y + f u = g.
$$
memungkinkan kita untuk mengubah persamaan umum ini menjadi
$$
    P(X,Y) = a X^2 + 2b X \cdot Y + c Y^2 + \cdots
$$

yang terlihat seperti polinomial umum orde dua.

Terminologi untuk klasifikasi PDE berasal dari fakta bahwa persamaan di atas menggambarkan irisan kerucut (conic sections) umum yang sering diklasifikasikan sebagai eliptik, parabolik, atau hiperbolik berdasarkan diskriminannya
$$
    b^2 - a c.
$$

Kemudian kita mengklasifikasikan PDE sesuai dengan tanda dari diskriminan tersebut sehingga

- Jika $b^2 - a c < 0$ → **Eliptik**
- Jika $b^2 - a c = 0$ → **Parabolik**
- Jika $b^2 - a c > 0$ → **Hiperbolik**


Alasan mengapa kita peduli terhadap klasifikasi adalah karena sering kali terdapat kelas metode numerik yang bekerja dengan baik untuk satu klasifikasi tetapi tidak untuk yang lainnya. Kita akan menyelidiki lebih lanjut mengapa dan bagaimana hal ini bekerja saat kita memperkenalkan setiap jenis PDE, tetapi untuk saat ini mari kita periksa beberapa contoh dasar PDE dari masing-masing klasifikasi.

## Contoh
**Persamaan Laplace:** Dalam kasus berdimensi 𝑛, persamaan Laplace adalah
$$
    \Delta u = \nabla^2 u = \sum^n_{i=1} u_{x_i x_i} = 0
$$
dan ketika 𝑛 = 2 dengan menggunakan 𝑥 dan 𝑦 sebagai dimensinya, kita peroleh
$$
    u_{xx} + u_{yy} = 0.
$$

Jenis PDE apakah ini?

Dengan mengidentifikasi koefisien dari persamaan di atas sebagai 
𝑎 = 1, 𝑏 = 0, dan 𝑐 = 1, maka diskriminannya menunjukkan bahwa persamaan Laplace bersifat eliptik karena
$$
    b^2 - a c = -1.
$$

**Persamaan Panas (Heat Equation):** Persamaan panas didefinisikan dalam satu dimensi ruang sebagai
$$
    u_t = u_{xx}.
$$
Jenis PDE apa persamaan panas ini?

Persamaan panas bersifat parabolik. Hal ini dapat dilihat dengan mengganti turunan terhadap 𝑦 di atas dengan 𝑡, sehingga diperoleh koefisien 𝑎 = −1, 𝑏 = 0, dan 𝑐 = 0, yang berarti diskriminannya sama dengan nol.

Bagaimana dengan persamaan panas dalam beberapa dimensi ruang?
$$
    u_t = u_{xx} + u_{yy}.
$$
Sayangnya, identifikasi tersebut tidak dapat langsung dilakukan di sini karena kita memiliki tiga variabel.

**Persamaan Gelombang (Wave Equation):** Persamaan gelombang dalam satu dimensi ruang didefinisikan sebagai
$$
    u_{tt} - u_{xx} = 0.
$$
Melalui proses eliminasi, Anda mungkin dapat menebak bahwa persamaan ini bersifat hiperbolik. Dengan kembali mengganti turunan terhadap 𝑦 dengan 𝑡, diperoleh 𝑎 = −1, 𝑏 = 0, dan 𝑐 = 1, sehingga
$$
    b^2 - a c = -(-1)(1) = 1.
$$

**Persamaan Navier–Stokes:** Memodelkan aliran fluida tak termampatkan
$$\begin{aligned}
    u_t + (u^2)_x + (uv)_y = - \frac{1}{\rho} p_x + \nu (u_{xx} + v_{yy})\\
    v_t + (uv)_x + (v^2)_y = - \frac{1}{\rho} p_y + \nu (u_{xx} + v_{yy})
\end{aligned}$$

Bahkan jika kita mengabaikan fakta bahwa persamaan ini tidak dapat diklasifikasikan secara langsung dengan tata nama kita, persamaan ini mengandung bagian yang menyerupai komponen hiperbolik (ruas kiri) dan komponen eliptik (ruas kanan), atau parabolik karena adanya turunan waktu. Oleh karena itu, persamaan Navier–Stokes sering dianggap bersifat parabolik, namun pendekatan penyelesaiannya sering menggunakan teknik dari setiap klasifikasi.

Salah satu cara yang lebih berguna untuk memahami klasifikasi PDE secara umum adalah dengan melihat bagaimana informasi merambat dalam suatu domain.

- PDE Hiperbolik: informasi merambat dengan kecepatan hingga (kecepatan grup bernilai nyata dan terbatas).

- PDE Parabolik: menggambarkan pergerakan informasi dengan kecepatan tak hingga (misalnya pada difusi dalam vakum).

- PDE Eliptik: merupakan kasus batas antara persamaan hiperbolik dan eliptik dalam hal perambatan informasi.

### 2. Analisis Fourier

Analisis Fourier adalah salah satu alat dasar yang sering digunakan untuk memahami dan menyelesaikan PDE linear.

Untuk daerah (domain) yang terbatas, biasanya kita menggunakan deret Fourier untuk menuliskan atau merepresentasikan solusi. Artinya, solusi tersebut diuraikan menjadi penjumlahan fungsi-fungsi sederhana (seperti sinus dan cosinus).

Penggunaan deret Fourier ini sebenarnya sama dengan menganggap fungsi-fungsi eigen dari operator diferensial sebagai dasar (basis) dalam menyusun solusi. Cara pandang ini berasal dari konsep aljabar linear, dan sangat kuat dalam membantu kita memahami serta menyelesaikan PDE.

Nanti konsep ini akan sering kita gunakan kembali.

### Fourier Series

Deret Fourier suatu fungsi memiliki beberapa bentuk, tergantung pada jenis fungsi yang sedang kita bahas. Perlu diingat bahwa 
sin 𝑥
sinx merupakan fungsi ganjil, sedangkan 
cos 𝑥
cosx merupakan fungsi genap. Pada interval yang terbatas, jika fungsi yang kita miliki adalah fungsi ganjil, maka kita dapat menggunakan deret Fourier sinus. Jika fungsi tersebut genap, maka kita dapat menggunakan deret Fourier cosinus. Namun, apabila fungsi tersebut tidak termasuk ganjil maupun genap, maka kita menggunakan deret Fourier lengkap, yang biasanya dituliskan dalam bentuk umum.
$$
    f(x) = a_0 + \sum^\infty_{n=1} a_n \cos \left( \frac{n \pi x}{L} \right) + \sum^\infty_{n=1} b_n \sin \left( \frac{n \pi x}{L} \right),
$$
Dimana:
$$\begin{aligned}
    a_0 &= \frac{1}{L} \int^L_0 f(x) dx, \\
    a_n &= \frac{2}{L} \int^L_0 f(x) \cos \frac{n \pi x}{L} dx, \\
    b_n &= \frac{2}{L} \int^L_0 f(x) \sin \frac{n \pi x}{L} dx.
\end{aligned}$$
Koefisien-koefisien tersebut merupakan proyeksi fungsi 𝑓(𝑥) ke dalam basis Fourier yang terdiri dari fungsi cosinus dan sinus, yang membentuk basis ortogonal (saling tegak lurus dalam arti integral).

Lalu, bagaimana hubungannya dengan aljabar linear dan ruang vektor berdimensi hingga?

### Fourier Transform

Transformasi Fourier dari sebuah fungsi $v(x)$ didefinisikan oleh

$$
\hat{v}(\xi) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} v(x) e^{-i\xi x} \, dx.
$$

Agar transformasi tersebut dapat eksis, kita memerlukan bahwa fungsi $v \in L^2$, yaitu

$$
\|v\|_2 = \left( \int_{-\infty}^{\infty} |v(x)|^2 \, dx \right)^{1/2} < \infty.
$$

Fungsi $\hat{v}(\xi) \in L^2$ karena dari identitas Perseval kita memiliki

$$
\|v(x)\|_2 = \|\hat{v}(\xi)\|_2.
$$

Kita juga dapat menggunakan transformasi invers untuk mendapatkan kembali fungsi asal:

$$
v(x) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} \hat{v}(\xi) e^{i\xi x} \, d\xi.
$$

Salah satu sifat paling penting dari Transformasi Fourier yang sudah kita gunakan di atas adalah bahwa transformasi Fourier mengubah turunan spasial menjadi hubungan aljabar:

$$
\hat{v}_x = i\xi \hat{v}.
$$

Hal ini merupakan hasil dari fungsi-fungsi

$$
e^{i\xi x}
$$

yang merupakan fungsi eigen dari operator

$$
\frac{\partial}{\partial x}
$$

Sifat ini juga berlaku pada kasus diskret, di mana kita akan menemukan bahwa analog diskret dari fungsi-fungsi eigen tersebut akan menjadi vektor eigen dari operator beda hingga yang setara.

**Latihan:** Cobalah ambil transformasi Fourier dari persamaan adveksi:

$$
u_t + c u_x = 0.
$$

Karena turunan terhadap waktu tidak dipengaruhi oleh transformasi Fourier, kita memperoleh

$$
\hat{u}_t = -i\xi c \hat{u}
$$

yang merupakan sistem tak hingga persamaan diferensial biasa (ODE). Sekarang cobalah selesaikan sistem ini untuk menemukan solusi dari persamaan adveksi.

Kita tahu bahwa solusi dari ODE-ODE tersebut berbentuk

$$
\hat{u}(\xi, t) = \hat{u}(\xi, 0) e^{-i\xi c t}.
$$

Dengan menggunakan transformasi invers, kita memperoleh

$$\begin{aligned}
    u(x,t) &= \frac{1}{\sqrt{2 \pi}} \int^\infty_{-\infty} \widehat{u}(\xi, 0)  e^{-i \xi c t} e^{i \xi x} d\xi, \\
    &= \frac{1}{\sqrt{2 \pi}} \int^\infty_{-\infty} \widehat{u}(\xi, 0)  e^{i \xi (x - c t)} d\xi, \\
    &= u(x - c t, 0).
\end{aligned}$$

## 3. Persamaan Eliptik

Persamaan eliptik paling sering muncul dalam bentuk **masalah nilai batas** (boundary value problems) pada suatu daerah tertentu.  

Sebelumnya kita telah menyebutkan persamaan Laplace, yang merupakan kasus khusus dari persamaan eliptik klasik lainnya, yaitu **masalah Poisson**:

$$
    \nabla^2 u = f.
$$

Jenis masalah ini sering ditemui ketika kita ingin mencari keadaan **setimbang** (steady-state) dari suatu masalah yang bergantung pada waktu, misalnya persamaan panas.  

Persamaan semacam ini juga dapat muncul dalam masalah yang bergantung pada waktu, di mana terdapat dinamika tertentu yang berlangsung **jauh lebih cepat** dibandingkan skala waktu dari bagian masalah lainnya. Salah satu contoh klasik adalah **kondisi inkompresibilitas** yang digunakan dalam persamaan Navier-Stokes (berupa masalah Poisson untuk tekanan).

Perilaku esensial dari masalah nilai batas eliptik adalah bahwa **solusi di bagian dalam daerah** $\Omega$ **ditentukan sepenuhnya** oleh **kondisi batas** dan fungsi $f$.  

Hal ini menjadi sangat penting ketika mempertimbangkan **metode numerik** untuk jenis masalah ini, karena kita sering kali harus menyelesaikan sistem persamaan linier yang **besar** dan **sparse** (jarang), di mana **setiap titik** di dalam daerah secara matematis **terhubung** (coupled) dengan **setiap titik lainnya**.

Secara umum, suatu **operator linier** $L$ berbentuk

$$
    L = \sum_{i,j=1}^n A_{ij} \frac{\partial^2}{\partial x_i \partial x_j} + \sum_{i=1}^n B_i \frac{\partial}{\partial x_i} + C
$$

dapat disebut **eliptik** jika matriks $A$ bersifat **definit positif** atau **definit negatif** (yaitu, $v^T A v$ memiliki tanda yang sama untuk setiap $v \neq 0$ dan $v \in \mathbb{R}^n$).

### Menyelesaikan Persamaan Eliptik

Sebagai contoh penyelesaian persamaan diferensial parsial (PDE) eliptik, mari kita pertimbangkan **persamaan Laplace** pada daerah persegi panjang:

$$\begin{aligned}
    &u_{xx} + u_{yy} = 0 \quad \Omega = (0, L) \times (0, H) \\
    &u(x,0) = f_1(x) \\
    &u(x,H) = f_2(x) \\
    &u(0,y) = g_1(y) \\
    &u(L,y) = g_2(y).
\end{aligned}$$

Karena kita memiliki **kondisi batas non-homogen** pada setiap sisi persegi panjang, kita **tidak dapat langsung** menggunakan metode pemisahan variabel. Sebagai gantinya, kita memecah masalah menjadi **empat sub-masalah** sehingga

$$
    u(x,y) = u_1(x,y) + u_2(x,y) + u_3(x,y) + u_4(x,y)
$$

di mana masing-masing $u_i(x,y)$ mempertahankan **satu sisi non-homogen** dari persegi panjang, sementara ketiga sisi lainnya diatur bernilai nol.

Sebagai contoh, untuk $u_1(x,y)$ kita memiliki:

$$\begin{aligned}
    \nabla^2 u_1 = 0 \quad & \quad \Omega = (0, L) \times (0, H) \\
    u_1(x,0) &= f_1(x) \\
    u_1(x,H) &= 0 \\
    u_1(0,y) &= 0 \\
    u_1(L,y) &= 0.
\end{aligned}$$

Menggunakan **pemisahan variabel**, kita mengasumsikan solusi berbentuk $u_4(x,y) = h(x) \phi(y)$.  
Ketika disubstitusikan ke dalam persamaan diferensial parsial (PDE), kita memperoleh:

$$\begin{aligned}
    &\phi \frac{\text{d}^2 h}{\text{d} x^2} + h \frac{\text{d}^2 \phi}{\text{d} y^2} = 0 \\
    &\frac{1}{h} \frac{\text{d}^2 h}{\text{d} x^2} = - \frac{1}{\phi} \frac{\text{d}^2 \phi}{\text{d} y^2} = -\lambda \quad \Rightarrow \\
    &h''(x) + \lambda h(x) = 0 \qquad \phi''(y) - \lambda \phi(y) = 0.
\end{aligned}$$

Kondisi batas untuk kedua fungsi adalah

$$
    h(0) = h(L) = 0 \quad \phi(0) = f_1(x) \quad \phi(H) = 0.
$$

Pertama, kita selesaikan masalah nilai batas homogen (BVP) yang lebih mudah untuk $h(x)$. Kita memperoleh

$$
    h(x) = A \cos (\sqrt{\lambda} x) + B \sin(\sqrt{\lambda} x)
$$

dengan menerapkan kondisi batas, kita dapatkan:

$$\begin{aligned}
    &h(0) = 0 = A \\
    &h(L) = 0 = B \sin(\sqrt{\lambda} L).
\end{aligned}$$

Untuk memenuhi kondisi batas tersebut, kita memiliki dua pilihan:  
Kita bisa mengasumsikan $B = 0$, yang akan menghasilkan **solusi trivial**, atau

$$
    \sqrt{\lambda} L = n \pi.
$$

Dengan menggunakan syarat yang kedua, kita memperoleh **nilai eigen** (eigenvalues)

$$
    \lambda_n = \left( \frac{n \pi}{L} \right)^2, \quad n = 1,2,\ldots
$$

sehingga solusi-solusi yang sesuai adalah

$$
    h_n(x) = B_n \sin\left( \frac{n \pi x}{L} \right).
$$

Untuk persamaan $\phi$, kita memiliki solusi berbentuk

$$
    \phi(y) = A e^{\sqrt{\lambda} y} + B e^{-\sqrt{\lambda} y}
$$

atau secara ekuivalen

$$
    \phi(y) = A \cosh(\sqrt{\lambda} y) + B \sinh(\sqrt{\lambda} y).
$$

Untuk memenuhi **kondisi batas homogen**, kita perlu memodifikasi solusi tersebut sedikit agar bergeser, sehingga menjadi

$$
    \phi(y) = A \cosh \left( \sqrt{\lambda} (y - H) \right) + B \sinh \left( \sqrt{\lambda} (y - H) \right)
$$

Bentuk ini memungkinkan kita memenuhi kondisi batas homogen dengan cara berikut:

$$\begin{aligned}
    \phi(H) = 0 
    &= A \cosh \left( \sqrt{\lambda} (H - H) \right) + B \sinh \left( \sqrt{\lambda} (H - H) \right) \\
    &= A \cosh (0) + B \sinh (0) \\
    &= A \cdot 1 + B \cdot 0 \\
    &= A
\end{aligned}$$

Sehingga $A = 0$, dan kita tinggal mendapatkan

$$
    \phi(y) = B \sinh \left( \frac{n \pi}{L} (y - H) \right)
$$

Dengan menggabungkan kedua solusi tersebut, kita memperoleh

$$
    u_{1,n}(x,y) = B_n \sin\left( \frac{n \pi x}{L} \right) \sinh \left( \frac{n \pi}{L} (y - H) \right)
$$

dan dengan menggunakan prinsip **superposisi**, solusi untuk sub-masalah $u_1$ menjadi

$$
    u_1(x,y) = \sum_{n=1}^\infty B_n \sin\left( \frac{n \pi x}{L} \right) \sinh \left( \frac{n \pi}{L} (y - H) \right)
$$

Sekarang, untuk memenuhi **kondisi batas terakhir** dan menentukan koefisien $B_n$, kita memiliki

$$
    u_1(x,0) = f_1(x) = \sum_{n=1}^\infty -B_n \sin\left( \frac{n \pi x}{L} \right) \sinh \left( \frac{n \pi H}{L} \right)
$$

dan dengan memanfaatkan **sifat ortogonalitas** fungsi-fungsi $\sin$, kita memperoleh:

$$\begin{aligned}
    \int_0^L f_1(x) \sin\left( \frac{m \pi x}{L} \right) \, dx 
    &= \sum_{n=1}^\infty -B_n \sinh \left( \frac{n \pi H}{L} \right) \int_0^L \sin\left( \frac{n \pi x}{L} \right) \sin\left( \frac{m \pi x}{L} \right) \, dx \\
    \int_0^L f_1(x) \sin\left( \frac{m \pi x}{L} \right) \, dx 
    &= -B_m \cdot \frac{L}{2} \sinh \left( \frac{m \pi H}{L} \right) \\
    B_m &= \frac{-2}{L \sinh \left( \frac{m \pi H}{L} \right)} \int_0^L f_1(x) \sin\left( \frac{m \pi x}{L} \right) \, dx.
\end{aligned}$$

### 4. Persamaan Parabolik

Persamaan parabolik memiliki kemiripan dengan persamaan eliptik, namun sering kali juga melibatkan ketergantungan terhadap waktu (meskipun tidak selalu demikian, tetapi hal ini akan menjadi fokus pembahasan kita). Sebagai contoh, kita dapat mengambil sembarang operator linear eliptik seperti yang telah didefinisikan sebelumnya, lalu membentuk sebuah persamaan parabolik dengan menuliskannya sebagai

$$
    u_t - Lu = f.
$$

Contoh klasik dari persamaan parabolik adalah persamaan panas (atau difusi), yang didefinisikan sebagai
$$
    u_t = \nabla^2 u.
$$
Persamaan diferensial parsial (PDE) ini memerlukan kondisi awal $u(x, 0)$ serta kondisi batas pada daerah $\Omega$.

### Menyelesaikan Persamaan Parabolik

Serupa dengan masalah eliptik, kita kembali akan menggunakan deret Fourier untuk menyelesaikan persamaan parabolik. Sekali lagi, kita pertimbangkan masalah model kita, yaitu persamaan panas, dalam satu dimensi ruang:
$$\begin{aligned}
    &u_t = \kappa u_{xx}, \quad \Omega = (0, T) \times (0, L), \\
    &u(x, 0) = f(x), \\
    &u(0, t) = 0, \quad \text{dan} \\
    &u(L, t) = 0.
\end{aligned}$$

Dengan menggunakan pemisahan variabel, kita mengasumsikan $u(x,t) = G(t) \phi(x)$ dan kemudian dapat menuliskan dua ODE berikut:
$$\begin{aligned}
    &\frac{\text{d} G}{\text{d} t} = -\kappa \lambda G, \\
    &\phi''(x) + \lambda \phi(x) = 0.
\end{aligned}$$
ODE yang bergantung pada waktu memiliki solusi dalam bentuk
$$
    G(t) = C e^{-\lambda \kappa t}.
$$
BVP untuk $\phi$ memiliki solusi dalam bentuk
$$
    \phi(x) = A \cos \left( \sqrt{\lambda} x \right) + B \sin \left( \sqrt{\lambda} x \right).
$$


Dengan menggunakan kondisi batas, kita menemukan bahwa $A = 0$ dan untuk solusi non-trivial,
$$
    \lambda = \left( \frac{n \pi}{L} \right )^2,
$$
yang menghasilkan solusi dalam bentuk
$$
    u_n(x,t) = B_n e^{-\left(\frac{n \pi}{L}\right)^2 \kappa t} \sin \left( \frac{n \pi x}{L} \right),
$$
dan dengan superposisi kita mendapatkan
$$
    u(x,t) = \sum^\infty_{i=1} B_n e^{-\left(\frac{n \pi}{L}\right)^2 \kappa t} \sin \left( \frac{n \pi x}{L} \right).
$$

Dengan mengevaluasi kondisi awal, kita memiliki
$$
    u(x,0) = f(x) = \sum^\infty_{i=1} B_n \sin \left( \frac{n \pi x}{L} \right),
$$
yang memberikan nilai $B_n$  dengan kembali menggunakan ortogonalitas fungsi $\sin$ :
$$\begin{aligned}
    \int^L_0 f(x) \sin \left( \frac{m \pi x}{L} \right) dx &= \sum^\infty_{i=1} B_n \int^L_0\sin \left( \frac{n \pi x}{L} \right) \sin \left( \frac{m \pi x}{L} \right) dx, \\
    \int^L_0 f(x) \sin \left( \frac{m \pi x}{L} \right) dx &= B_m \frac{L}{2}, \\
    B_m &= \frac{2}{L} \int^L_0 f(x) \sin \left( \frac{m \pi x}{L} \right) dx,
\end{aligned}$$
sehingga PDE terselesaikan.

## 5. Persamaan Hiperbolik

Persamaan hiperbolik sering muncul sebagai hukum kekekalan atau hukum keseimbangan. Sebagai contoh, kekekalan momentum (Hukum Kedua Newton) merupakan salah satu bentuk hukum keseimbangan. Demikian pula, kekekalan energi dapat dituliskan sebagai persamaan hiperbolik. Persamaan hiperbolik klasik tentu saja adalah persamaan gelombang

$$
u_{tt} - c^2 u_{xx} = 0.
$$

Namun, dalam banyak kasus ketika membahas metode numerik untuk persamaan hiperbolik, kita sering menguraikan PDE orde dua menjadi sistem PDE orde satu, misalnya

$$
u_t + A u_x = 0,
$$

dengan $A$ adalah sebuah matriks. Dalam hal ini, PDE dikatakan hiperbolik jika nilai eigen $\lambda$ dari matriks $A$ adalah real dan matriks tersebut dapat didiagonalisasi.


Jika kita mengambil turunan penuh (atau turunan total) dari solusi baru tersebut, kita peroleh

$$
\frac{\text{d} u}{\text{d} t} = \frac{\partial u}{\partial t} + \frac{\text{d} x}{\text{d} t} \frac{\partial u}{\partial x},
$$

yang setelah mentransformasikan persamaan awal dengan menggunakan $\xi$, membawa kita pada kesimpulan bahwa PDE semula dapat dituliskan kembali menjadi dua ODE:

$$
\begin{aligned}
    &\frac{\text{d} u}{\text{d} t} = 0, \quad \text{dan}\\
    &\frac{\text{d} x}{\text{d} t} = c(x,t,u).
\end{aligned}
$$

Sekarang kita mengetahui bahwa $u(x(t),t)$ konstan sepanjang kurva $x(t)$, dan secara khusus

$$
u(x,t) = u(x(t), 0).
$$

Dengan demikian, penyelesaian PDE bergantung pada pencarian kurva karakteristik yang diberikan oleh $x(t)$.

Sebagai contoh sederhana, mari kita selesaikan persamaan adveksi linear pada domain tak hingga

$$
\begin{aligned}
    &u_t + c u_x = 0, \quad \Omega = (-\infty, \infty) \times (0, T) \\
    &u(x,0) = f(x), \quad \text{dan} \\
    &u_x(\pm \infty, t) \rightarrow 0.
\end{aligned}
$$

Persamaan karakteristiknya adalah

$$
\frac{\text{d} x}{\text{d} t} = c,
$$

yang solusinya adalah $x(t) = c t + x_0$.  

Dengan demikian, kita dapat menggambarkan karakteristik sebagai garis-garis lurus

![Constant Wave Speed](./images/characteristics.png)

dengan persamaan garis-garis tersebut adalah $x(t) = c t + x_0$.

Untuk persamaan nonlinier, yaitu ketika kecepatan bergantung pada solusi $c(u)$, kondisi awal akan menentukan kemiringan karakteristik sehingga menghasilkan dua kemungkinan kasus.  

Pertimbangkan persamaan Burger

$$
u_t + u u_x = 0,
$$

dengan pengaturan yang sama seperti PDE linear sebelumnya. Cobalah tentukan kurva karakteristik untuk kondisi awal berikut

$$
u(x,0) = \left \{ \begin{aligned}
        4, \;\; & x<0, \\
        3, \;\; & x>0.
    \end{aligned} \right.
$$

Apa bentuk karakteristik untuk permasalahan ini? Masalah apa yang kemudian muncul dalam kasus ini?

Kurva karakteristiknya adalah

$$
x(t) = \left \{ \begin{aligned}
        4 t + x_0, \;\; & x<0, \\
        3 t + x_0, \;\; & x>0,
    \end{aligned} \right.
$$

Karakteristik-karakteristik tersebut akan saling bertumbukan (collision), sehingga menyebabkan terbentuknya shock (gelombang kejut).

![Shock](./images/characteristics_shock.png)

Kita dapat menemukan persamaan untuk lokasi gelombang kejut (shock) dengan menggunakan kondisi Rankine–Hugoniot:

$$
\frac{\text{d} x_s}{\text{d} t} = \frac{f(x_+, t) - f(x_-, t)}{u_+ - u_-},
$$

di mana $x_s(t)$ adalah posisi shock, $f$ adalah fungsi fluks yang didefinisikan dari persamaan $u_t + f(u)_x = 0$ , dan notasi $u_\pm$ menyatakan nilai di sebelah kiri dan kanan shock, tergantung pada tandanya.

Bagaimana jika kita memiliki kondisi awal

$$
    u(x,0) = \left \{ \begin{aligned}
        3, \;\; & x<0, \\
        4, \;\; & x>0,
    \end{aligned} \right.
$$

sebagai gantinya? Jenis solusi apa yang sekarang kita harapkan?

Kurva karakteristik sekarang adalah

$$
    x(t) = \left \{ \begin{aligned}
        3 t + x_0, \;\; & x<0, \\
        4 t + x_0, \;\; & x>0,
    \end{aligned} \right. 
$$

yang menyebabkan terdapat suatu bagian pada bidang $(x,t)$ yang tidak dilalui oleh karakteristik mana pun.

Sebagai gantinya, kita mengasumsikan solusi yang berubah secara mulus, yang disebut rarefaksi (rarefaction), untuk mengisi bagian solusi yang hilang tersebut.

![Rarefaction](./images/characteristics_rarefaction.png)



Dalam kasus ini, kita ingin mencari solusi yang berubah secara mulus di dalam daerah berbentuk irisan (wedge) tersebut. Kita melakukannya dengan mengasumsikan bahwa solusinya bersifat self-similar.

Hal ini berarti bahwa $u$ berbentuk

$$
u(x,t) = u(\xi),
$$

di mana untuk PDE hiperbolik orde satu, variabel self-similar adalah

$$
\xi = \frac{x}{t}.
$$

Cobalah mensubstitusikan bentuk ini ke dalam persamaan asli. Hal tersebut akan memberikan solusi dalam bentuk $u(\xi)$ untuk bagian dalam daerah irisan (wedge) tersebut.

$$\begin{aligned}
    &u_t + u u_x = 0 \\
    &u' \frac{\partial \xi}{\partial t} + u u' \frac{\partial \xi}{\partial x} = 0 \\
    &u' \left ( -\frac{x}{t^2} + u \frac{1}{t} \right ) = 0 \\
    &u' \left ( -\xi + u\right ) = 0
\end{aligned}$$

Persamaan terakhir ini sekarang dapat diselesaikan.

Karena hasilnya adalah

$$
u'(\xi),(-\xi + u) = 0,
$$

maka terdapat dua kemungkinan:

1. $u'(\xi) = 0$, yang berarti $u$ konstan, atau
2. $-\xi + u = 0$, sehingga diperoleh
   $
   u = \xi.
   $

Untuk solusi rarefaksi di bagian dalam daerah irisan (wedge), kita mengambil solusi tak konstan, yaitu

$$
u(x,t) = \xi = \frac{x}{t}.
$$

Inilah solusi yang berubah secara mulus di dalam daerah rarefaksi.
